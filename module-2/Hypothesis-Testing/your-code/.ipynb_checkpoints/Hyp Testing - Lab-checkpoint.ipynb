# Challenge 2 - Hypothesis Tests

In this section of the lab, we will test whether the hourly wage of all hourly workers is significantly different from $30/hr. Import the correct one sample test function from scipy and perform the hypothesis test for a 95% two sided confidence interval.

```python
# so here we would need to pull out the H1 and H0
```

```python
# So here we should take a look at the annual salary of each worker. 

salary_data = employee_data[employee_data['Annual Salary'].isna()]
salary_data = salary_data.drop(['Annual Salary'], axis=1)
```

```python
# lets have a look at what I did here. 

salary_data
```

```python
#okay, first thing that needs to happen here, is that we need to change the $ and have it not be in the column any longer.
#it's not recognized 
salary_data['Hourly Rate'] = salary_data['Hourly Rate'].str.replace("[$,]", "", regex=True)
salary_data['Hourly Rate'] = salary_data['Hourly Rate'].astype(float)
salary_data.head()
```

```python
#now that the $ is gone, we can move on to the testing
```

```python
from scipy.stats import ttest_1samp

stat, pval = ttest_1samp(salary_data['Hourly Rate'], 30)
print('stat is  ', stat)
print('pvalue for the two-tailed test is ', pval)
```

```python
np.mean(salary_data)
```

```python
#so considering that the pvalue is pretty low, i'm going to assume that the average salary is not $30 hourly.
#we will have to reject the H0. When we calculate the mean we can see that the average hourly salary is $33.
```

# Challenge 3: To practice - Constructing Confidence Intervals

While testing our hypothesis is a great way to gather empirical evidence for accepting or rejecting the hypothesis, another way to gather evidence is by creating a confidence interval. A confidence interval gives us information about the true mean of the population. So for a 95% confidence interval, we are 95% sure that the mean of the population is within the confidence interval. 
).

To read more about confidence intervals, click [here](https://en.wikipedia.org/wiki/Confidence_interval).


In the cell below, we will construct a 95% confidence interval for the mean hourly wage of all hourly workers. 

The confidence interval is computed in SciPy using the `t.interval` function. You can read more about this function [here](https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.t.html).

To compute the confidence interval of the hourly wage, use the 0.95 for the confidence level, number of rows - 1 for degrees of freedom, the mean of the sample for the location parameter and the standard error for the scale. The standard error can be computed using [this](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.sem.html) function in SciPy.

```python
import scipy.stats

confidence_level = 0.95
degrees_freedom = len(salary_data['Hourly Rate']) - 1
sample_mean = np.mean(salary_data['Hourly Rate'])
sample_standard_error = scipy.stats.sem(salary_data['Hourly Rate'])


confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```

Now construct the 95% confidence interval for all salaried employeed in the police in the cell below.

```python
confidence_level = 0.95
degrees_freedom = len(police_salary_data['Annual Salary']) - 1
sample_mean = np.mean(police_salary_data['Annual Salary'])
sample_standard_error = scipy.stats.sem(police_salary_data['Annual Salary'])


confidence_interval = scipy.stats.t.interval(confidence_level, 
                                             degrees_freedom, 
                                             sample_mean, 
                                             sample_standard_error)
```

```python
print( 'confidence interval is ', confidence_interval, '.' )
```
